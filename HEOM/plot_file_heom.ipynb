```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib.lines import Line2D
from matplotlib.legend_handler import HandlerTuple
```

# Exciton 300K

```python
mash_exc_300 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/mash_exc300K.dat")
heom_exc_300 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/pop_exc_tau-50_T-300_init-0_K_2_L_2.dat")

fig, ax = plt.subplots()

# Create empty lists to store handles for legend
handles = []
labels = []

for i in range(1, 8):
    line_mash, = ax.plot(mash_exc_300[:, 0], mash_exc_300[:, i], ":", label='MASH')
    line_heom, = ax.plot(heom_exc_300[:, 0], heom_exc_300[:, i], label='HEOM')

    if i == 1:
        handles.extend([line_mash, line_heom])
        labels.extend(['MASH', 'HEOM'])

# Create legends using handles and labels
l = ax.legend(handles, labels, handler_map={tuple: HandlerTuple(ndivide=None)})

plt.xlim(0,5000)
plt.xlabel("Time (fs)")
plt.ylabel("Population")
plt.show()
```
