# Exciton 300 K Optimization   

```python
resid1 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/exc_300/scan_0_5000_exc300_resid1.dat")
resid2 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/exc_300/scan_0_5000_exc300_resid2.dat")
resid3 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/exc_300/scan_0_5000_exc300_resid3.dat")

time = 400
index_t = np.where(resid1[:,0]==time)[0]
print(index_t)


plt.plot(resid1[:,0],resid1[:,1], label = "Full Train")
plt.plot(resid2[:,0],resid2[:,1], label = "Extrapolation")
plt.plot(resid3[:,0],resid3[:,1], label = "Half Train")

plt.xlabel('t (fs)')
plt.ylabel('RMSE')
plt.legend()

# Add an 'X' marker at the point
plt.plot(resid3[74,0],resid3[74,1], marker='x', markersize=10, label='Annotated Point', color='black')
print(resid3[74,0])

# Add a text label
plt.text(resid3[74,0]-40,resid3[74,1]+0.0006, '400 fs', fontsize=10, color='black')

plt.title("Exciton, T=300K, One Initial State")
plt.show
```

```python
from optimizer_optimated import Optimizer
kappa_exc_300 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/exc_300/scan_0_5000_exc300_kappas.dat")
heom_exc_300 = np.loadtxt("/u/dem/kebl6911/Part-II/HEOM/pop_exc_tau-50_T-300_init-0_K_2_L_2.dat")

optimizer = Optimizer(heom_exc_300,heom_exc_300[-1,1:8],7)
t1s = 5*np.arange(6,400)

times = heom_exc_300[heom_exc_300[:, 0] > t1s[74], 0]
p0 = heom_exc_300[heom_exc_300[:, 0] == t1s[74], 1:8]

predict = optimizer.predict(p0, times,"forward", kappa_exc_300[74])
print(np.shape(predict))

fig, ax = plt.subplots()
# Create empty lists to store handles for legend
handles = []
labels = []

for i in range(1, 8):
    c = "C%i"%i
    line_heom, = ax.plot(heom_exc_300[:, 0], heom_exc_300[:, i], label='HEOM', color=c)
    line_opt, = ax.plot(times, predict[:, i-1], "D", label='optimized', marker='D', markersize=3, markevery=250, color=c)

    if i == 1:
        handles.extend([line_opt, line_heom])
        labels.extend(['Optimized', 'HEOM'])

# Create legends using handles and labels
l = ax.legend(handles, labels, handler_map={tuple: HandlerTuple(ndivide=None)})

plt.axvline(times[0], linestyle=':', color="black")
plt.text(times[0]-120, -0.11, '400', va='bottom', ha='left')

plt.xlim(0,5000)
plt.xlabel("Time (fs)")
plt.ylabel(r'$P_{\mathrm{1\rightarrow n}}$')
plt.title("Exciton, T=300K, t=400 fs")
plt.show()
```
